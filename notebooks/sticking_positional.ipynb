```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("full_error_data_extracted.csv", index_col=0)
df.head()
```

```python
df.columns
```

```python
exclude_cols = ['id', 'displayed_text', 'label_annotations', 'span_annotations',
       'behavioral_data', 'annotator_id',"Sticking with the wrong diagnosis", "Cannot pick any error category", "Reasonable response by GPT-4"]
cols_to_sum = df.columns.difference(exclude_cols)
```

```python
df["sum_nonstick"] = df[cols_to_sum].sum(axis=1)
df.head()
```

```python
df_stick = df[df["Sticking with the wrong diagnosis"] == 1]
df_stick = df_stick[df_stick["sum_nonstick"] >= 1]
df_stick.head()
```

### insight
Sticking error has been annotated in 193 cases and co-occurs with another class in 64.77% of them, while occuring in 108 unique questions.

### sticking error position relative to other errors

```python
positions = {"first":0,
             "middle":0,
             "last": 0}
pos_per_cat = dict()

for col in cols_to_sum:
    temp = df_stick.filter([col, "Sticking with the wrong diagnosis", "span_annotations"])
    temp = temp[temp[col]==1]
    first, last, middle = 0,0,0
    tgt_cols = [col, "Sticking with the wrong diagnosis"]
    for idx, row in temp.iterrows():
        spans = eval(row["span_annotations"])
        spans = [d for d in spans if d["annotation"] in (tgt_cols)]
        if spans[0]["annotation"].startswith("Sticking"):
            first += 1
        elif spans[-1]["annotation"].startswith("Sticking"):
            last += 1
        else:
            middle +=1
    pos_per_cat[col] = {"first": first,
                        "first_perc": first / len(temp),
                        "middle": middle,
                        "middle_perc": middle / len(temp),
                        "last": last,
                        "last_perc": last / len(temp)}
    positions["first"] += first
    positions["last"] += last
    positions["middle"] += middle
    print(len(temp), first+last+middle)


print(positions)
pos_per_cat
```

#### sticking error position overall

```python
positions = {"first":0,
             "middle":0,
             "last": 0}


first, last, middle = 0,0,0
for idx, row in df_stick.iterrows():
    spans = eval(row["span_annotations"])
    if spans[0]["annotation"].startswith("Sticking"):
        first += 1
    elif spans[-1]["annotation"].startswith("Sticking"):
        last += 1
    else:
        middle +=1

positions["first"] += first
positions["last"] += last
positions["middle"] += middle
print(len(df_stick), first+last+middle)


print(positions)
for k,v in positions.items():
    print(k, v/len(df_stick))
```

### insight

When the sticking error co-occurs, it usually occurs a first (60%) or before other errors (first+middle = 73.6%). However, this is dependant on which class it co-occurs with: when labeled with error classes that co-occur often ($n_{ij}>.33) with this error, it appears first in at least $73.\overline{33}$ of cases. In contrast, for errors with lower co-occurence rates, this behaviour is not apparant or less pronounced.
