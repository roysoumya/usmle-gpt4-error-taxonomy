# Experience EDA

```python
pth1 = "./combined_data_11thJan2024.csv"
df_annot = pd.read_csv(pth1)
df_annot = df_annot[df_annot["annotator_id"].isin(approved_ids)]
df_combined = df_annot
df_annot.head()
```

```python
survey_ans = df_combined[df_combined.id.str.contains("html")]
feedback_a = survey_ans[survey_ans.id.str.contains("experience")].to_dict()
demo = survey_ans[survey_ans.id.str.contains("demographics")].to_dict()

def remove_shells(_dict):
    k = list(_dict["label_annotations"].keys())[0]
    return eval(_dict["label_annotations"][k])
```

```python
def extract_answers_one_row(_dict):
    # extracts all questions and answers of one response
    # ie. one experience/demographics questionnaire
    keys = list(_dict.keys())
    _res = {}
    for k in keys:
        ans = list(_dict[k].keys())[0]
        if ans.find("text_box")!=-1 or ans.find("select-one")!=-1:
            res = _dict[k][ans]
        else:
            res = ans
        _res[k] = res
    return _res


def extract_and_append_to_list(df, tgt_lst_exp, tgt_lst_demo):
    survey_ans = df[df.id.str.contains("html")]
    for index, row in survey_ans.iterrows():
        # read demographic or experience feedback
        res = extract_answers_one_row(eval(row["label_annotations"]))
        res["annotator_id"] = row["annotator_id"]
        if row["displayed_text"] == "experience":
            tgt_lst_exp.append(res)
        else:
            tgt_lst_demo.append(res)        

           
exp_list, demo_list = [], []
extract_and_append_to_list(df_combined, exp_list, demo_list)
        
```

```python
len(exp_list), len(demo_list)
```

```python
df_exp = pd.DataFrame(exp_list).infer_objects()
df_demo = pd.DataFrame(demo_list).infer_objects()

df_exp.to_csv("experience_approved_only.csv")
df_demo.to_csv("demographics_approved_only.csv")

len(df_exp), len(df_demo)
```
